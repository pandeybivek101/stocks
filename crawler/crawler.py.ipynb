```
def f_crawl_company_data(url:str)->list:

    """
    Crawl the required information from web page

    Args:
        url (string ): url of the site to be crawled
        
    Returns:
        items (list): list of tuples with required attributes
    """
    page = requests.get(url)
    html = BeautifulSoup(page.text, "html.parser")
   
    items=[]
    for panel in html.find('div', id='accordion').findAll('div', attrs={'class':'panel panel-default'}):
        panel_title_tag=panel.div.h3.a
        panel_title = panel_title_tag.text
        try:
            for index, tr in enumerate(panel.findAll('table')[0].findAll('tr')):
                if index != 0:
                    company_code_tag=tr.td.a
                    company_name_tag=company_code_tag.find_parent().findNextSibling()
                    listed_share_tag=company_name_tag.findNextSibling()
                    paid_value_up_tag=listed_share_tag.findNextSibling()
                    total_paid_up_tag=paid_value_up_tag.findNextSibling()

                    company_tuple=(panel_title, company_code_tag.text, 
                                company_name_tag.text, listed_share_tag.text.strip(),
                                paid_value_up_tag.text.strip(), total_paid_up_tag.text.strip()
                                )
                    
                    
                    items.append(company_tuple)

        except:
            print(f"something went wrong while scrapping {panel_title}")
                
    return items


url="https://merolagani.com/CompanyList.aspx"
print(f_crawl_company_data(url))
           
```
